## Synthetic Daily Data

```python
import pandas as pd
import numpy as np

daily = pd.read_csv('Daily_data.csv')
daily = daily.rename(columns={'AdjClose': 'real_daily'})
daily
```

### Raw Price Comparison

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style = 'white', palette = 'pastel')
fig, ax = plt.subplots(figsize=(8,6))
daily.plot(ax=ax, title='BRK-B',secondary_y='synthetic_daily', style=['-', '--'],lw=1)
```

```python
idx = np.random.randint(252-48)

dataframe = pd.DataFrame({'Real': daily['real_daily'][idx:idx+48], 'Synthetic': daily['synthetic_daily'][idx:idx+48]})
dataframe.plot(figsize=(8,6),title='BRK-B',secondary_y='Synthetic', style=['-', '--'],lw=1)
```

```python
idx = np.random.randint(252-48)

dataframe = pd.DataFrame({'Real': daily['real_daily'][idx:idx+48], 'Synthetic': daily['synthetic_daily'][idx:idx+48]})
dataframe.plot(figsize=(8,6),title='BRK-B',secondary_y='Synthetic', style=['-', '--'],lw=1)
```

### Log Return Comparison

```python
real_daily_return = np.log(daily['real_daily']/daily['real_daily'].shift())
synthetic_daily_return = np.log(daily['synthetic_daily']/daily['synthetic_daily'].shift())
```

```python
sns.set_theme(style = 'white', palette = 'pastel')
fig, ax = plt.subplots(figsize=(8,6))
real_daily_return[1:].plot(ax=ax, label = 'Real Data')
synthetic_daily_return[1:].plot(ax=ax, label = 'Synthetic Data')
ax.set_xlabel('Log return')
ax.set_ylabel('Freq of log return')
ax.set_title('BRK-B Daily Real vs. Synthetic Data Volatility')
ax.legend(['Real Data','Synthetic Data'])
plt.show()
```

```python
sns.set_theme(style = 'white', palette = 'pastel')
fig, ax = plt.subplots(figsize=(8,6))
synthetic_daily_return[-44:].plot(ax=ax, label = 'Synthetic Data')
real_daily_return[-44:].plot(ax=ax, label = 'Real Data')
ax.set_xlabel('Log return')
ax.set_ylabel('Freq of log return')
ax.set_title('BRK-B Daily Real vs. Synthetic Data Volatility')
ax.legend(['Synthetic Data','Real Data'])
plt.show()
```

### Volatility Clustering

```python
sns.set_theme(style = 'white', palette = 'pastel')
fig, ax = plt.subplots(figsize=(8,6))
real_daily_return[1:].hist(ax=ax, label = 'Real Data')
ax.set_xlabel('Log return')
ax.set_ylabel('Freq of log return')
ax.set_title('BRK-B Daily Real Data Volatility')
plt.grid(b=None)
plt.legend()
plt.show()
```

```python
sns.set_theme(style = 'white', palette = 'pastel')
fig, ax = plt.subplots(figsize=(8,6))
real_daily_return[1:].hist(ax=ax, label = 'Real Data')
synthetic_daily_return[1:].hist(ax=ax, alpha=0.6, label = 'Synthetic Data')
ax.set_xlabel('Log return')
ax.set_ylabel('Freq of log return')
ax.set_title('BRK-B Daily Real vs. Synthetic Data Volatility')
plt.grid(b=None)
plt.legend()
plt.show()
```
